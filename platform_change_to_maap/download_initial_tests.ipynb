```python

# ============================================================
# Árbol: collection -> productType -> version -> url de ejemplo
# ============================================================
# Estrategia:
#   Para cada collection, busca hasta max_items_per_collection items
#   con platform = 'EarthCARE' y construye el árbol con los valores
#   extraídos de properties (productType, version) y el asset href.
# ============================================================

from collections import defaultdict
import json

MAX_PER_COLLECTION = 300  # suficiente para cubrir variedad de tipos/versiones

catalog_tree = {}   # { collection: { productType: { version: [url_ejemplo] } } }
errors = {}

for collection in EC_COLLECTIONS:
    try:
        search = catalog.search(
            collections=[collection],
            filter="platform = 'EarthCARE'",
            method="GET",
            max_items=MAX_PER_COLLECTION,
        )
        items = list(search.items())
    except Exception as e:
        errors[collection] = f"search error: {e}"
        continue

    if not items:
        catalog_tree[collection] = {}
        continue

    tree = defaultdict(lambda: defaultdict(list))

    for item in items:
        props    = item.properties or {}
        pt       = props.get("productType") or props.get("product:type") or "?"
        version  = props.get("version") or "?"

        # Toma el primer href disponible como URL de ejemplo
        href = None
        for _, asset in item.assets.items():
            href = getattr(asset, "href", None)
            if href:
                break

        # Solo guarda un ejemplo por (productType, version) para no repetir
        if not tree[pt][version]:
            tree[pt][version] = href or "— no href —"

    catalog_tree[collection] = {pt: dict(vers) for pt, vers in tree.items()}

# -------- Impresión en árbol --------
print("=" * 72)
print("ÁRBOL: EarthCARE collections -> productType -> version -> url")
print("=" * 72)

for collection, types in catalog_tree.items():
    status = f"({len(types)} productType(s))" if types else "(sin datos)"
    print(f"\n📁 {collection}  {status}")
    if not types:
        continue
    for pt, versions in sorted(types.items()):
        print(f"  ├── productType: {pt}")
        verslist = sorted(versions.items())
        for i, (ver, url) in enumerate(verslist):
            prefix = "  │   └──" if i == len(verslist) - 1 else "  │   ├──"
            print(f"{prefix} version: {ver}")
            print(f"  │       url: {url}")

if errors:
    print("\n⚠️  Errores:")
    for c, msg in errors.items():
        print(f"  {c}: {msg}")

print("\n--- dict completo (para copiar/pegar) ---")
print(json.dumps(catalog_tree, indent=2, ensure_ascii=False))
```
